```python
import cv2
import numpy as np
import os
from time import time
import supervision as sv
import matplotlib.pyplot as plt
```

```python


num_test = 6 
#Test haar cascade classifier
def test_image():
    #Create matplotlib subplot equals to num_test
    fig, axs = plt.subplots(2, num_test//2)
    axs = axs.ravel()
    for i in range(num_test):
        id = np.random.randint(1, 700)
        img_path = "positive/y{id}.jpg".format(id=id)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        tumor_cascade = cv2.CascadeClassifier('models/cascade.xml')
        tumors = tumor_cascade.detectMultiScale(gray)
        for (x,y,w,h) in tumors:
            cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)
        
        title = f"{len(tumors)} Tumors Detected"
        axs[i].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        axs[i].axis('off')
        axs[i].set_title(title)
    plt.show()
        

test_image()
 
```
